```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
import pandas as pd

w = 64
h = 64

for fn in os.listdir('Data/Raw'):
    print(fn)
    name = fn.replace('.png','')
    img = cv2.imread(f'Data/Raw/{fn}')
    if img is not None:
        df = pd.read_csv(f'Data/CSVs/{name}.csv')
        
        img_w, img_h, _ = img.shape
        bb = []
        for index, row in df.iterrows():
            if not 'Total carotid' in row['name'] and not 'Lipid plaque' in row['name']:
                # Extract all of the image's bounding boxes
                ymin = int(row['ymin'])
                ymax = int(row['ymax'])
                xmin = int(row['xmin'])
                xmax = int(row['xmax'])
                bb.append((xmin, ymin, xmax, ymax))

        n = 0
        for y in range(0, img_h - h, h):
            for x in range(0, img_w - w, w):
                inside = False
                for box in bb:
                    # If the 64x64 intercept a bounding box 
                    if ((x >= box[0] and x <= box[2]) or (x+w >= box[0] and x+w <= box[2])) and ((y >= box[1] and y <= box[3]) or (y+h >= box[1] and y+h <= box[3])):
                        inside = True
                        break
                if inside:
                    continue
                temp_img = img[y:y+h, x:x+w]

                file_location_image = f'Data/Patches/NoPlaque/sagittal_{name}_noplaque_{n}.png'
                cv2.imwrite(file_location_image, temp_img)
                #plt.imshow(temp_img)
                #plt.show()
                #print(temp_img.shape)
                n += 1
        print(n)
    
```
